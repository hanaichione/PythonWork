### 1) 연습코드

```python
##### 시 목록 가져오기

# si_list_raw = driver.find_element_by_xpath('//*[@id="SIDO_NM0"]') # 셀렉트박스에 접근
# si_list = si_list_raw.find_elements_by_tag_name("option")
si_list_raw = driver.find_element(By.XPATH, '//*[@id="SIDO_NM0"]')
si_list = driver.find_elements(By.TAG_NAME, "option")
si_list[4].get_attribute("value")

si_names = [op.get_attribute("value") for op in si_list]
si_names.remove("")
si_names
```

```python
from tqdm import tqdm_notebook
import time

for gu in tqdm_notebook(gu_names):
    tag = driver.find_element(By.ID, "SIGUNGU_NM0")
    tag.send_keys(gu)
    
    time.sleep(1)
    
    driver.find_element(By.XPATH, '//*[@id="searRgSelect"]/span').click()
    
    time.sleep(1)
    
    driver.find_element(By.XPATH, '//*[@id="glopopd_excel"]/span').click()
    
    time.sleep(1)
```

## (2) Data Definition

```python
from glob import glob
```

```python
# 해당하는 파일명을 가져와 리스트로 저장
station_files = glob("data/oil/지역*.*")

tmp_raw = []
for f in station_files:
    tmp = pd.read_excel(f, header=2)
    tmp_raw.append(tmp)
```

```python
len(tmp_raw)
```

```python
##### 데이터 프레임으로 합치기
station_raw = pd.concat(tmp_raw)
station_raw.head(50)
```

```python
station_raw.info()
```

### (3) Data Preprocessing

원하는 컬럼만 추출 : 상호, 주소, 상표, 셀프여부, 휘발유

```python
stations = pd.DataFrame({"Oil_Store":station_raw["상호"], "주소":station_raw["주소"],
             "가격":station_raw["휘발유"], "셀프":station_raw["셀프여부"], 
              "상표":station_raw["상표"]})
stations
```

```python
##### 구 파생변수 추가

stations["구"] = [gu.split()[1] for gu in stations["주소"]]
```

```python
stations.head(50)
```

```python
print(stations["구"].unique())
print(len(stations["구"].unique()))
```

```python
##### 가격 결측치 확인

stations["가격"].unique() # '-' 로 표시
stations[stations["가격"] == '-'] # 결측치 레코드는 삭제
stations = stations[stations["가격"] != '-']
```

```python
stations.info()
```

```python
##### 가격 정보를 숫자로 변환

stations["가격"] = stations["가격"].astype(float)
stations.info()
```

```python
##### 인덱스 재조정

stations.reset_index(inplace=True) # drop=True
```

```python
del stations["index"]
```

```python
stations.head()
```
