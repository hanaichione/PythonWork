```python
##### total_bill에 대한 금액 분포 확인 : 데이터가 하나뿐, 히스토그램 이용

fig = plt.figure() # 그래프를 만들 저장 공간
ax = fig.add_subplot(1, 1, 1) # nrows, ncols, index

ax.hist(tips["total_bill"], bins=20)
ax.set_title("식사 대금에 대한 히스토그램")
ax.set_xlabel("frequency")
ax.set_ylabel("Total Bill")
```

```python
##### total_bill과 tip의 관계

fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)

ax.scatter(tips["total_bill"], tips.tip)
ax.set_xlabel("total_bill")
ax.set_ylabel("tip")
```

```python
##### 성별과 팁의 관계
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)

ax.boxplot([tips[tips["sex"]=="Male"]["tip"], 
          tips[tips["sex"]=="Female"]["tip"]], labels=["남성", "여성"])
```

```python
##### 성별에 따른 total_bill과 tip의 관계

def record_sex(sex):
    if sex == "Female":
        return 0
    else:
        return 1
    
tips["sex_color"] = tips["sex"].apply(record_sex)
tips.head(10)

fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)

ax.scatter(tips["total_bill"], tips["tip"], c=tips["sex_color"], alpha=0.7, s=tips["size"]*30) 
# c는 컬러 설정, alpha는 투명도 설정, s는 컬럼 크기(여기서는 size)에 따라 점의 크기 다르게 표시하기
```

### (1) Data Prepare

```python
train = pd.read_csv("data/train.csv")
train.head(10)
```

```python
train.info()
```

```python
train.groupby("Pclass")[["Survived"]].count()
```

### (2) EDA(Explore Data Analysys)

```python
def bar_chart(feature):
    survived = train[train["Survived"]==1][feature].value_counts()
    dead = train[train["Survived"]==0][feature].value_counts()
    
    df = pd.DataFrame([survived, dead])
    df.index = ["Survived", "Dead"]
    df.plot(kind="bar", stacked=True)
```

```python
bar_chart("Sex")
```

```python
bar_chart("Pclass")
```

```python
bar_chart("SibSp")
```

```python
bar_chart("Parch")
```

```python
bar_chart("Embarked") # 탑승한 동네 : 부자 동네는 1등급에 타서 많이 살지 않았을까
```

```python
train.info()
```

```python
##### Name에서 호칭 부분 추출하기 : 정규표현식 활용

train["title"] = train["Name"].str.extract(" (\S+)[.]")
train["title"].value_counts()
```

```python
bar_chart("title")
```

```python
title_mapping = {"Mr":0, "Miss":1, "Mrs":2, "Master":3, "Dr":3, "Rev":3, "Major":3, "Mlle":3,
                "Col":3, "Sir":3, "Don":3, "Mme":3, "Lady":3, "Johkheer":3, "Countess":3,
                "Capt":3, "Ms":3}
# Mr 0 Miss 1 Mrs 2 그외 3

train["title"] = train["title"].map(title_mapping) 
train["title"].value_counts()
```

```python
train["Age"].fillna(train.groupby("title")["Age"].
                    transform("mean"), inplace=True) # transform 사용하여 모든 행에 다 적용
train.info()
```

```python
bar_chart("title")
```

```python
facet = sns.FacetGrid(train, hue="Survived", aspect=3) # 분포와 함께 정보를 알 수 있는 선 그래프(확률밀도함수)
# FaceGrid는 기존 그래프를 활용하는 것이기 때문에 그래프, 기준이 되는 변수를 인자로 줘야 함

facet.map(sns.kdeplot, "Age", shade=True)
facet.add_legend() # legend : 범례
facet.set(xlim=(0, train["Age"].max())) # x축 눈금 조정 등 다양한 속성 조정 가능
```

```python
facet = sns.FacetGrid(train, hue="Survived", aspect=3)

facet.map(sns.kdeplot, "Age", shade=True)
facet.add_legend() # legend : 범례
facet.set(xlim=(10, 30))
```

```python
txt = open("data/wordcloud/constitution.txt")
txt

# read : 전체 문자열로
# readlines : 전체를 읽어오되 한줄을 리스트의 원소로
# readline : 한줄을 읽어옴

text = txt.read()
text
```

```python
?WordCloud.generate # 빈도수 세서 정렬해 줌
```

```python
wc = WordCloud().generate(text) # WordCloud는 객체, 객체 생성 후 함수 호출
print(dir(wc))
wc.words_ # 빈도수별로 정렬까지 완료
```

```python
plt.figure(figsize=(12, 12))
plt.imshow(wc)
```

```python
wc = WordCloud(max_font_size=50, background_color="white").generate(text)
plt.figure(figsize=(12, 12))
plt.imshow(wc, interpolation="bilinear") # imshow : 숫자를 이미지로 변환, interpolation="bilinear" : 계단현상 완화
plt.axis("off") # x,y축 눈금 제거
```

```python
import PIL

alice_mask = np.array(PIL.Image.open("data/wordcloud/alice_mask.png")) # 이미지를 배열로 바꾸기
# 이미지는 3차원, 여러장이면 4차원 (가로 * 세로 * 색상수)
# 영상은 5차원 (이미지 * 프레임)

alice_mask.shape

plt.figure(figsize=(8, 8))
plt.imshow(alice_mask) # 계단현상 완화
plt.axis("off") # x,y축 눈금 제거
```

```python
text = open("data/wordcloud/alice.txt").read()

stopwords = set(STOPWORDS)
stopwords.add("said")

wc = WordCloud(max_font_size=60, background_color="white", 
              stopwords=stopwords, max_words=100, mask=alice_mask).generate(text) # 불용어 정의, 단어 수 조절 가능

plt.figure(figsize=(12,12))
plt.imshow(wc) 
plt.axis("off") # x,y축 눈금 제거
```

```python
STOPWORDS # 불용어사전, set 형태, set은 중복을 불허
```

```python
text = open("data/wordcloud/alice.txt").read()

stopwords = set(STOPWORDS)
stopwords.add("said")
alice_color = np.array(PIL.Image.open("data/wordcloud/alice.png"))

wc = WordCloud(max_font_size=60, background_color="white", 
              stopwords=stopwords, max_words=200, mask=alice_color).generate(text) # 불용어 정의, 단어 수 조절 가능

image_color = wordcloud.ImageColorGenerator(alice_color)

plt.figure(figsize=(12,12))
plt.imshow(wc.recolor(color_func=image_color)) 
plt.axis("off") # x,y축 눈금 제거
```

### 한글로 워드클라우드 만들기
- pip install konlpy
- https://konlpy.org

#### import module/object

```python
from konlpy.tag import Okt
from konlpy.corpus import kolaw
from wordcloud import WordCloud
import nltk # 빈도수 세기 위한 모듈
```

```python
##### 파일 읽어오기
ko_text = kolaw.open("constitution.txt").read()
ko_text
```

```python
##### Okt 사용, 명사 추출
okt = Okt()
okt.nouns("아버지가방에들어가신다.")
token_ko = okt.nouns(ko_text) # 명사 토큰 추출
print(len(token_ko))
```

```python
token_ko
```

```python
##### 불용어 처리
stop_words = ["제", "월", "일", "조", "수", "때", "그", "이", "바", "및", "안", "정", "위", "관"]
```

```python
token_ko = [each_word for each_word in token_ko if each_word not in stop_words]
print(len(token_ko)) # 3882 -> 3211
```

```python
?nltk
```

```python
ko = nltk.Text(token_ko, name="대한민국 헌법")
print(dir(ko))
data = ko.vocab().most_common(500) # 가장 빈도수가 높은 원소 100개
wc = WordCloud(max_font_size=40, background_color="white", max_words=300,
              font_path="C:\\Windows\\Fonts\\malgun.ttf")\
                .generate_from_frequencies(dict(data)) # generate_from_frequencies는 인자로 dict

plt.figure(figsize=(12,12))
plt.imshow(wc, interpolation="bilinear") 
plt.axis("off") # x,y축 눈금 제거
```
