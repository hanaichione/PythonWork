```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

import matplotlib.font_manager as fm
font_name = fm.FontProperties(fname="C:\\Windows\\Fonts\\malgun.ttf").get_name()
plt.rc("font", family=font_name)

import matplotlib as mlp
mlp.rcParams["axes.unicode_minus"] = False

import urllib.request as req
from urllib.request import urlopen
from urllib.error import HTTPError, URLError
import urllib.parse
from bs4 import BeautifulSoup
import bs4
import requests

import json
import re
```

```python
# json 파일 생성
data = {}
data["data"] = []

# 멤버별로 json 파일 다르게 생성
# 이윤석 0-999.json
# 송지민 1000-1999.json
# 홍진우 2000-2999.json
# 전우진 3000-3999.json
with open('0-999data.json', 'w', encoding="utf-8") as make_file:    ####################################### 수정해야할 곳 ####################
    json.dump(data, make_file, ensure_ascii=False, indent="\t")
```

```python
# 웹 소스 준비
url = "https://2bob.co.kr/recipe.php?id=list&eTheme=1"      ####################################### 수정해야할 곳 ####################
# page = urlopen (url)

header = {"User-Agent" : "Mozilla/5.0"}
page = requests.get(url, headers=header)
page
# 가져온 웹 소스를 트리 구조로 변환
soup = BeautifulSoup(page.text, "html.parser")
```

```python
# 페이지당 4줄, 4줄의 맨 앞에있는 레시피만 ml_none 클래스 부여되어있음
# 따라서 뒤에 따라오는 4개의 레시피는 해당 레시피 기준 siblings을 사용하여 가져옴

# 가로기준 맨 앞 항목의 링크 가져오기
urls = soup.select('li.ml_none > a')
lead_urls_list = []
for url in urls:
    lead_urls_list.append(url["href"])

# 맨 앞 항목의 링크를 기준으로 따라오는 4개의 링크 가져오기
follow_url_list=[]
for lead_url in lead_urls_list:
    urls = soup.find("a", href=lead_url).parent.next_siblings
    for url in urls:
        if(type(url) == bs4.element.Tag):
            follow_url_list.append(list(url.children)[1]["href"])

# 두개의 리스트 합치기
url_list = lead_urls_list + follow_url_list

# 합쳐진 리스트의 중복 제거
url_list = set(url_list)
url_list = list(url_list)
print(url_list)
print(len(url_list))

# 괄호, \r\n 제거를 위한 패턴 생성
pattern1 = r'\([^)]*\)'
pattern2 = '\r\n'

# 생성해 놓은 json파일 load
with open('0-999data.json', "r", encoding='utf-8') as json_file:    ####################################### 수정해야할 곳 ####################
    json_data = json.load(json_file)

# urljoin을 위한 base 선언
base = "https://2bob.co.kr"

# 각자의 url별로 크롤링 실행
# cnt 값 매번 조정 필요
# 이윤석 0-999
# 송지민 1000-1999
# 홍진우 2000-2999
# 전우진 3000-3999
cnt = 0                         ####################################### 수정해야할 곳 ####################
for i in url_list:
    # 웹 소스 준비
    url = urllib.parse.urljoin(base, i)

    header = {"User-Agent" : "Mozilla/5.0"}
    page = requests.get(url, headers=header)
    page
    # 가져온 웹 소스를 트리 구조로 변환
    soup = BeautifulSoup(page.text, "html.parser")

    # 레시피 명
    title = soup.find('h2', class_="prod_title").get_text()
    print(title)

    # 필수재료
    ingredients = soup.find('p', class_="mate_list").get_text()
    
    # 재료명에서 괄호와 괄호안의 정보 삭제
    ingredients = re.sub(pattern=pattern1, repl='', string=ingredients)

    # 재료명에서 \r\n 제거
    ingredients = re.sub(pattern=pattern2, repl='', string=ingredients)

    # 리스트로 변환
    temp_list = ingredients.split(',')

    ingredients_list = []
    # 공백 제거
    for i in temp_list:
        i = re.sub(" ", "", string=i)
        ingredients_list.append(i)

    print(ingredients_list)

    json_data['data'].append({
        "code" : cnt,
        "title" : title,
        "ingredients" : ingredients_list,
        "url" : url
    })
    print("--------------------------------------------------")
    cnt = cnt + 1
# json 파일 저장
with open('0-999data.json', "w", encoding='utf-8') as outfile:      ####################################### 수정해야할 곳 ####################
    json.dump(json_data, outfile, indent="\t", ensure_ascii=False)
    
```
