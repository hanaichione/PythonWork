## 데이터베이스 추가 하기 위해 재료 분할

```python
ingred_raw = pd.read_excel('Deduplicated_Filtered_Ingredients_List_up_by_value_counts.xlsx', header=None)
```

```python
ingred_raw
```

```python
ingred = ingred_raw[0]
ingred
```

```python
ingredients1 = []
ingredients2 = []
ingredients3 = []
ingredients4 = []
```

```python
for i in range(len(ingred_raw[0])):
    if(i%4 == 0):
        ingredients1.append(ingred[i])
    if(i%4 == 1):
        ingredients2.append(ingred[i])
    if(i%4 == 2):
        ingredients3.append(ingred[i])  
    if(i%4 == 3):
        ingredients4.append(ingred[i])    
```

```python
ingredients1=pd.DataFrame(ingredients1)
ingredients2=pd.DataFrame(ingredients2)
ingredients3=pd.DataFrame(ingredients3)
ingredients4=pd.DataFrame(ingredients4)
```
