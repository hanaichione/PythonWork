```python
import json
import pandas as pd
```

```python
with open('recipe_data.json', encoding="utf-8") as f:
    data = json.load(f)
```

```python
data
```

```python
data['data']
```

```python
# json 파일 생성
data1 = {}
data1["data"] = []

for d in range(len(data['data'])):
    if(d%4 == 0):
        print("출력")
        data1["data"].append(data['data'][d])
```

```python
data1
```

```python
with open('recipe_data_1.json', "w", encoding='utf-8') as outfile:    
                    json.dump(data1, outfile, indent="\t", ensure_ascii=False)
```

```python
ingredients1 = []
ingredients2 = []
ingredients3 = []

for i in range(len(ingred)):
    if(i<=960):
        if(i%3 == 0):
            ingredients1.append(ingred[0][i])
        if(i%3 == 1):
            ingredients2.append(ingred[0][i])
        if(i%3 == 2):
            ingredients3.append(ingred[0][i])
```
